**Limpieza de datos del dataset-4**

Se van a identificar si existen valores nulos y/o filas o columnas vacías así como información no relevante para el análisis de datos del tema seleccionado.

```python
#Importamos las librerías necesarias
import pandas as pd
import numpy as np
```

```python
#Cargamos la base de datos a limpiar
file = 'suicidio.csv'

#Leemos el archivo
df = pd.read_csv(file)
```

```python
#Visualizamos las columnas de la base de datos para ver la forma
#que esta tiene.

df.head()
```

```python
#Viendo mejor las columnas
df.columns
```

```python
#Renombramos las columnas:
df = df.rename(columns={'Region Name':'Region',
                        'Country Name':'País',
                        'Year':'Años',
                       'Sex':'Sexo',
                       '0 Year':'0 años',
                       '1-4 Years': '1-4 Años',
                       '5-9 Years': '5-9 Años',
                       '10-14 Years': '10-14 Años',
                       '15-19 Years': '15-19 Años',
                       '20-24 Years': '20-24 Años',
                       '25-29 Years': '25-29 Años',
                       '30-34 Years': '30-34 Años',
                       '35-39 Years': '35-39 Años',
                       '40-44 Years': '40-44 Años',
                       '45-49 Years': '45-49 Años',
                       '50-54 Years': '50-54 Años',
                       '55-59 Years': '55-59 Años',
                       '60-64 Years': '60-64 Años',
                       '65-69 Years': '65-69 Años',
                       '70-74 Years': '70-74 Años',
                       '75-79 Years': '75-79 Años',
                       '80-84 Years': '80-84 Años',
                       '85+ Years': '+85 Años',
                       'Unknown Age': 'Edad desconocida',
                       'No of Suicides':'N° de suicidios',
                       'Percentage of cause-specific deaths out of total deaths': 'Suicidio/Muertes generales'})

df = df.rename(columns={'Death rate per 100 000 population': 'Tasa por 100000 habitantes'})
```

```python
#Resultando:
df.head()
```

```python
#Visualizamos la forma de los datos que vienen:
# shape = (# de datos, #número de columnas)
df.shape
```

```python
#Estudiamos la información del data-frame: esto nos dará el conteo y el tipo de datos a estudiar
#Esto nos dirá si existen datos faltantes o incompletos dependiendo de las características a estudiar
df.info()
```

```python
#De la misma forma podemos verificar si existen datos nulos mediante:
print(df.isnull().sum())
```

Ahora imputaremos las columnas que no nos interesan estudiar con el fin de reducir la data a estudiar y analizar posteriormente en el procesamiento y organización de los datos.

En el caso de esta base de datos serán:
La edad de 0 años y de 1-4 años y las edades desconocidas.

```python
#Imputamos datos (eliminamos columnas)
df = df.drop ('0 años', axis = 1)
df = df.drop ('1-4 Años', axis = 1)
df = df.drop ('Edad desconocida', axis = 1)
```

```python
#Luego mediante el método dropna verificamos y eliminamos filas con uno o más valores nulos con los atributos 'all' y 'any'
df.dropna(how='all') #Elimina filas completamente nulas
df.dropna(how='any') #Elimina filas que tenga al menos un valor nulo
```

```python
#Vemos que no se eliminan datos por lo que la base de datos es completa en todas sus entradas pero siempre deben reducirse
#con el fin de delimitar los datos de procesamiento y análisis.

#Verificando
df.info()
```

```python
#Eliminamos en caso de que existan, registros duplicados
df.drop_duplicates()
```

Como podemos ver hemos realizado una limpieza, o mejor dicho una verificación de la cantidad de datos y vemos que la base de datos no contiene datos nulos y/o vacíos.

```python
#Cambiando a español los registro de Sexo
genero_map = {
    'Male': 'Masculino',
    'Female': 'Femenino'
}

# Aplica la traducción a los registros de la columna ya renombrada
df['Sexo'] = df['Sexo'].replace(genero_map)
```

```python
df.head()
```

Ahora contemos la cantidad de registros de las distintas regiones por si hay que descartar algunos o si se tienen suficientes registros.

```python
#Contar registros que contienen "America" en la columna 'Region'
cuenta_america = df['Region'].str.contains('America', case=False, na=False).sum()
print(f"Cantidad de registros que contienen 'America': {cuenta_america}")

print("--------------------------")

#Contar registros que contienen "Oceania" en la columna 'Region'
cuenta_oceania = df['Region'].str.contains('Oceania', case=False, na=False).sum()
print(f"Cantidad de registros que contienen 'Oceania': {cuenta_oceania}")

print("--------------------------")

#Contar registros que contienen "Asia" en la columna 'Region'
cuenta_asia = df['Region'].str.contains('Asia', case=False, na=False).sum()
print(f"Cantidad de registros que contienen 'Asia': {cuenta_asia}")

print("--------------------------")

#Contar registros que contienen "África" en la columna 'Region'
cuenta_africa = df['Region'].str.contains('Africa', case=False, na=False).sum()
print(f"Cantidad de registros que contienen 'África': {cuenta_africa}")

print("--------------------------")

#Contar registros que contienen "Europa" en la columna 'Region'
cuenta_europa = df['Region'].str.contains('Europe', case=False, na=False).sum()
print(f"Cantidad de registros que contienen 'Europa': {cuenta_europa}")
```

Vemos que con las regiones se tienen la cantidad de datos suficientes para ser analizados y conseguir un patrón entre ellos

Ahora veamos los registros que existen en los distintos años que van desde 1950 hasta 2021, con el fin de descartar si existen pocos datos en ciertos años.

```python
#Primero desarrollemos el código para un año específico:
año_especifico = 2004
conteo_año_especifico = df[df['Años'] == año_especifico].shape[0]
print(f"El conteo de registros para el año {año_especifico} es: {conteo_año_especifico}")
```

```python
#Como hacer esto uno por uno lo vuelve un poco complicado, por lo que haremos un for que itere sobre los años desde 1950 a 2021
for año in range(1950, 2022):
    conteo_año = df[df['Años'] == año].shape[0]
    print(f"El conteo de registros para el año {año} es: {conteo_año}")
```

Esto último permite ver lo siguiente:
 - Enfoque por generación: boomers, X, Y, Z
 - Al tener un enfoque por generación y la edad del grupo de personas se pueden saber a qué generación pertenece e inferir información sobre ello. (Parte de análisis)
 - No se elimina directamente los registros con datos menores a 100 ya que al ser registros pequeños pueden procesarse rápidamente para ver tasas de crecimiento.
