Ahora con la data divida de tal forma podemos ver los rangos de edad y la respectiva generación a la que pertenecen. Es decir se procede a introducir una columna que catalogue esa información.

# Análisis data frame : df_ninez (Niñez y Preadolescencia)

```python
def asignar_generacion_sucesiva(anio, edad_min, edad_max):
    # Calcular el año de nacimiento para el rango de edad
    nacimiento_min = anio - edad_max
    nacimiento_max = anio - edad_min
    
    # Definir las generaciones
    generaciones = {
        'Grandiosos': (1900, 1927),
        'Silenciosos': (1928, 1945),
        'Boomers': (1946, 1964),
        'Generación X': (1965, 1981),
        'Generación Y': (1982, 1996),
        'Generación Z': (1997, 2012),
        'Generación Alfa': (2013, 2020)
    }

    # Lista para las generaciones correspondientes
    generacion_asignada = []
    
    # Recorrer las generaciones para ver a qué generaciones corresponde el rango de año
    generaciones_list = list(generaciones.items())
    for i in range(len(generaciones_list) - 1):
        generacion, (min_age, max_age) = generaciones_list[i]
        siguiente_generacion, (min_age_next, max_age_next) = generaciones_list[i + 1]
        
        # Verificar si el rango de edad de nacimiento cae dentro de las generaciones sucesivas
        if nacimiento_min >= min_age and nacimiento_max <= max_age:
            generacion_asignada.append(generacion)
        elif nacimiento_min >= min_age_next and nacimiento_max <= max_age_next:
            generacion_asignada.append(siguiente_generacion)
        elif nacimiento_min < min_age and nacimiento_max > max_age:
            generacion_asignada.append(f'{generacion}/{siguiente_generacion}')
    
    # Asignar generaciones sucesivas según el caso
    if not generacion_asignada:
        if nacimiento_min < 1928:  # Caso Grandiosos/Silenciosos
            generacion_asignada.append('Grandiosos/Silenciosos')
        elif nacimiento_min >= 1928 and nacimiento_min <= 1945:  # Caso Silenciosos/Boomers
            generacion_asignada.append('Silenciosos/Boomers')
        elif nacimiento_min >= 1946 and nacimiento_min <= 1964:  # Caso Boomers/Generación X
            generacion_asignada.append('Boomers/Generación X')
        elif nacimiento_min >= 1965 and nacimiento_min <= 1981:  # Caso Generación X/Generación Y
            generacion_asignada.append('Generación X/Generación Y')
        elif nacimiento_min >= 1982 and nacimiento_min <= 1996:  # Caso Generación Y/Generación Z
            generacion_asignada.append('Generación Y/Generación Z')
        elif nacimiento_min >= 1997 and nacimiento_min <= 2012:  # Caso Generación Z/Generación Alfa
            generacion_asignada.append('Generación Z/Generación Alfa')
        elif nacimiento_min >= 2013:  # Caso Generación Alfa
            generacion_asignada.append('Generación Alfa')

    return ', '.join(generacion_asignada)

# Crear las nuevas columnas basadas en los años y rangos de edad
df_ninez['G 5-9 Años'] = df_ninez['Años'].apply(lambda x: asignar_generacion_sucesiva(x, 5, 9))
df_ninez['G 10-14 Años'] = df_ninez['Años'].apply(lambda x: asignar_generacion_sucesiva(x, 10, 14))
```
