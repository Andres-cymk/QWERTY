```python
import pandas as pd
#estatus civil
df = pd.read_csv('master.csv')
df.drop('HDI for year', axis=1, inplace=True)
df.drop('country-year', axis=1, inplace=True)
# Eliminar filas con al menos un valor nulo y crear un nuevo dataframe
df_sin_filas_nulas = df.dropna()

# eliminar columnas con al menos un valor nulo 
df.dropna(axis=1, inplace=True)

# eliminar filas donde todos los valores son nulos
df.dropna(how='any')

print(df.head())
print("____________________________________________________________")
print("datos nulos")
print(df.isnull().sum())
print("_____________________________________________________________")
print("que columnas tienen los datos?")
print(df.columns)
print("_____________________________________________________________")
print("qué tamaño tienen los datos?")
print(df.shape)
print("_____________________________________________________________")
print("hay valores nulos en los datos?")
print(df.info())
print("_____________________________________________________________")
print("como se distribuyen las variables numéricas")
print(df.describe())
print("_____________________________________________________________")
print("como se comportan las variables categóricas")
print("_____________________________________________________________")
print(df.describe(include=['O']))
print(df.isnull())

```
